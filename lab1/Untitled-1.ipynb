```python
dane_uzytkownika = input("Podaj swoje imię i nazwisko: ").strip()

czesci = dane_uzytkownika.split()
imie = czesci[0]
nazwisko = czesci[-1] if len(czesci) > 1 else ""

if nazwisko:
    inicjaly = f"{imie[0].upper()}.{nazwisko[0].upper()}."
else:
    inicjaly = f"{imie[0].upper()}."
print(f"Inicjały: {inicjaly}")

imie_od_tylu = imie[::-1]
print(f"Imię od tyłu: {imie_od_tylu}")

liczba_liter = len(dane_uzytkownika.replace(" ", ""))
print(f"Liczba liter (bez spacji): {liczba_liter}")
```

```python
zakupy = [
    {"nazwa": "mleko", "cena": 4.50, "ilość": 2},
    {"nazwa": "chleb", "cena": 5.00, "ilość": 1},
    {"nazwa": "masło", "cena": 7.99, "ilość": 1},
    {"nazwa": "jajka", "cena": 12.50, "ilość": 1},
    {"nazwa": "mleko", "cena": 4.50, "ilość": 3},
]

laczny_koszt = sum(produkt["cena"] * produkt["ilość"] for produkt in zakupy)
print(f"1. Łączny koszt zakupów: {laczny_koszt:.2f} zł")

najdrozszy_produkt = max(zakupy, key=lambda p: p["cena"])
print(f"2. Najdroższy produkt: {najdrozszy_produkt['nazwa']} (cena: {najdrozszy_produkt['cena']:.2f} zł)")

unikalne_nazwy = list(set(produkt["nazwa"] for produkt in zakupy))
print(f"3. Unikalne nazwy produktów: {unikalne_nazwy}")

produkty_powyzej_5 = [produkt for produkt in zakupy if produkt["cena"] > 5.00]

print("4. Produkty droższe niż 5 zł:")
for p in produkty_powyzej_5:
    print(f"   - {p['nazwa']} ({p['cena']:.2f} zł)")
```

```python
import random

while True:
    wylosowana_liczba = random.randint(1, 100)
    liczba_prob = 0
    print("\nWylosowałem liczbę z zakresu od 1 do 100. Spróbuj ją odgadnąć!")
    
    while True:
        wprowadzony_tekst = input("Podaj liczbę: ")
        
        if not wprowadzony_tekst.isdigit():
            print("To nie jest poprawna liczba całkowita. Spróbuj jeszcze raz!")
            continue
            
        zgadywana_liczba = int(wprowadzony_tekst)
        liczba_prob += 1
        
        if zgadywana_liczba < wylosowana_liczba:
            print("Za mało!")
        elif zgadywana_liczba > wylosowana_liczba:
            print("Za dużo!")
        else:
            print(f"Brawo! Szukana liczba to {wylosowana_liczba}.")
            print(f"Zgadłeś za {liczba_prob} razem!")
            break
            
    odpowiedz = input("Czy chcesz zagrać ponownie? (t/n): ").strip().lower()
    
    if odpowiedz != 't' and odpowiedz != 'tak':
        print("Dziękuję za grę! Do zobaczenia.")
        break
```

```python
def policz_slowa(tekst):
    """Zwraca liczbę słów w tekście."""
    return len(tekst.split())

def najdluzsze_slowo(tekst):
    """Zwraca najdłuższe słowo w tekście."""
    slowa = [slowo.strip(".,!?;:()[]\"'") for slowo in tekst.split()]
    
    return max(slowa, key=len)

def czestotliwosc_slow(tekst):
    """Zwraca słownik {słowo: liczba_wystąpień}."""
    wystapienia = {}
    slowa = [slowo.strip(".,!?;:()[]\"'").lower() for slowo in tekst.split()]
    
    for slowo in slowa:
        if slowo:
            wystapienia[slowo] = wystapienia.get(slowo, 0) + 1
            
    return wystapienia

def cenzura(tekst, zakazane_slowa):
    """Zamienia zakazane słowa na '***'."""
    ocenzurowany_tekst = tekst
    for zakazane in zakazane_slowa:
        ocenzurowany_tekst = ocenzurowany_tekst.replace(zakazane, "***")
        ocenzurowany_tekst = ocenzurowany_tekst.replace(zakazane.capitalize(), "***")
        
    return ocenzurowany_tekst

test_tekst = """
Python - język programowania wysokiego poziomu, 
ogólnego przeznaczenia, o rozbudowanym pakiecie bibliotek standardowych, 
którego ideą przewodnią jest czytelność i klarowność kodu źródłowego. 
Jego składnia cechuje się przejrzystością i zwięzłością. 
To wspaniały i cudowny język programowania!
"""

print("--- ORYGINALNY TEKST ---")
print(test_tekst.strip())

print("\n--- 1. LICZBA SŁÓW ---")
print(f"Liczba słów: {policz_slowa(test_tekst)}")

print("\n--- 2. NAJDŁUŻSZE SŁOWO ---")
print(f"Najdłuższe słowo: '{najdluzsze_slowo(test_tekst)}'")

print("\n--- 3. CZĘSTOTLIWOŚĆ SŁÓW ---")
slownik_czestotliwosci = czestotliwosc_slow(test_tekst)
powtarzajace_sie = {k: v for k, v in slownik_czestotliwosci.items() if v > 1}
print(f"Słowa, które wystąpiły więcej niż 1 raz: {powtarzajace_sie}")

print("\n--- 4. CENZURA ---")
slownik_przeklenstw = ["cudowny", "wspaniały", "język"]
print(cenzura(test_tekst, slownik_przeklenstw).strip())
```
